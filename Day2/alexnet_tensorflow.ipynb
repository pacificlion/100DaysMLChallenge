Saturating functions : which squeeze the input R to bounded output:

Relu is non saturating as limit is [0,inf)

To saturate means to fill up to a point where no more can be added. In the context of a saturating function, it means that after a certain point, any further increase in the function's input will no longer cause a (meaningful) increase in its output, which has (very nearly) reached its maximum value. The function at that point is "all filled up", so to speak (or saturated).
